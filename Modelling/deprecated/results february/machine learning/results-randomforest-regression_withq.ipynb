```python
bacteria = pd.read_csv(loc_train)
```

```python
bacteria.head()
```

```python
bacteria.columns
print bacteria.columns
```

```python
bacteria_names = filter(lambda b: b[0] == 'b', bacteria.columns)     # Select all the bacteria names
pathogen_names = filter(lambda p: p[0:3] == 'LMG', bacteria.columns) # Select all the pathogen names
```

```python
feature_cols = bacteria_names + ['evenness', 'cellcount']
print feature_cols
```

```python
X = bacteria[feature_cols]
Y = bacteria[pathogen_names]
Y.head()
```

```python
test_indices = sample(bacteria.index, 40)
train_indices = [index for index in bacteria.index if index not in test_indices]
X_train = X.ix[train_indices]# Select the rows
Y_train = Y.ix[train_indices]# Select the rows
X_test = X.ix[test_indices]  # Select the rows
Y_test = Y.ix[test_indices]  # Select the rows
print X_test.head()
print Y_test.head()
```

## Training and predictions

```python
# Initialize random forest

clf = ensemble.RandomForestRegressor(n_estimators = 1000, n_jobs=-1)
clf.fit(X_train, Y_train)
predictions = clf.predict(X_test)
(np.mean((Y_test - predictions)**2))
```

```python
Y.var()
```

```python
print clf.feature_importances_
```

```python
n_range =  range(50, 1500, 50)
pred_error = []
MSE = 0
for n in n_range:
    regr = ensemble.RandomForestRegressor(n_estimators = n, n_jobs=-1)
    regr.fit(X_train, Y_train)
    predictions = regr.predict(X_test)
    MSE = np.mean((Y_test - predictions)**2)
    pred_error += [MSE]


    
```

```python
figure(figsize=(12, 9))

# Remove the plot frame lines. They are unnecessary chartjunk.  
ax = subplot(111)  
ax.spines["top"].set_visible(False)  
ax.spines["right"].set_visible(False) 

# Ensure that the axis ticks only show up on the bottom and left of the plot.  
# Ticks on the right and top of the plot are generally unnecessary chartjunk.  
ax.get_xaxis().tick_bottom()  
ax.get_yaxis().tick_left()  

# Make sure your axis ticks are large enough to be easily read.  
# You don't want your viewers squinting to read your plot.  
xticks(range(0,6), fontsize=22)  
yticks(fontsize=22)  

# Along the same vein, make sure your axis labels are large  
# enough to be easily read as well. Make them slightly larger  
# than your axis tick labels so they stand out.  
xlabel("Number of estimator trees", fontsize=22)  
ylabel("MSE", fontsize=22) 

title("MSE for estimating pathogen invasion", fontsize=26,  y=1.08)

plt.plot(n_range, pred_error, linewidth=6)
classes = Y_test.columns.values
plt.legend(classes,loc=4, fontsize= 20)
savefig("MSE_pathogeninvasion", bbox_inches="tight")
```
