```python
onlinetext = bs4.BeautifulSoup(res.text)
```

```python
selection =  onlinetext.select('li')
```

```python
selection
```

```python
names = []
links = []
titles = []
formats = []
```

```python
cd = os.listdir('.')
```

```python
selection[80].select("a")[0].get("href")
```

```python
last_name = [name.split(",")[0] for name in names]
first_name = [name.split(",")[1].split(";")[0] for name in names]
```

```python
for tag in selection:
    try:
        link = tag.select("a")[0].get("href")
        if ".pdf" in link:
            links.append(link)
            for i, j in enumerate(tag.strings):
                if i == 0:
                    names.append(unicode(j.string).encode('ascii','ignore'))
                elif i == 1:
                    titles.append(unicode(j.string).encode('ascii','ignore'))
                elif i == 2:
                    formats.append(unicode(j.string).encode('ascii','ignore'))
                    
    except:
        print tag
```

```python
save = [thesis.split("/")[-1] for thesis in links]
```

```python
len(names), len(titles), len(formats), len(links), len(save)
```

```python
database_theses = pd.DataFrame([first_name, last_name, titles, formats, links, save], index=["first_name", "last_name", "title", "format", "link", "save"]).T
```

```python
error = []
cd = os.listdir('./downloads')
for name, url in zip(save, links):
    try:
        if name not in cd:
            if 'http:' not in url:
                # url = 'http://www.labmet.ugent.be' + url
                url = 'www.labmet.ugent.be' + url
                print url
            webbrowser(url)
            rq = requests.get(url)
            playFile = open('downloads/' + name, 'wb')
            for chunk in rq.iter_content(100000):
                playFile.write(chunk)
            playFile.close()
    except:
        print url
        error.append(url)
```
