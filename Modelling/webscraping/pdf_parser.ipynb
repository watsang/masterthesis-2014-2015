```python
full_name = []
for row in xrange(database_theses.shape[0]):
    first_name = database_theses["first_name"][row][1:]
    last_name = database_theses["last_name"][row]
    first_name = first_name.replace('.', '')
    last_name = last_name.replace('.', '_')
    first_plus_last = "_".join([first_name, last_name])
    full_name.append(first_plus_last)
```

```python
database_theses['full_name'] = full_name
```
