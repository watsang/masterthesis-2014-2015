# Random Forest

## Exploring Dataset

```python
import pandas as pd
import numpy as np
from sklearn import ensemble
from sklearn.preprocessing import StandardScaler
from random import sample
import matplotlib.pyplot as plt
```

```python
if __name__ == "__main__":
    loc_train = 'C:\\Users\\Tsang\\Documents\\univ\\2014-2015\\thesis\\results february\\design_results_febr_counts.csv'
```

```python
bacteria = pd.read_csv(loc_train)
```

```python
bacteria.head()
```

```python
bacteria.columns
print bacteria.columns
```

```python
bacteria_names = filter(lambda b: b[0] == 'b', bacteria.columns)     # Select all the bacteria names
pathogen_names = filter(lambda p: p[0:3] == 'LMG', bacteria.columns) # Select all the pathogen names
```

```python
feature_cols = bacteria_names + ['evenness', 'cellcount']
print feature_cols
```

```python
X = bacteria[feature_cols]
Y = bacteria[pathogen_names]
Y.head()
```

```python
test_indices = sample(bacteria.index, 40)
train_indices = [index for index in bacteria.index if index not in test_indices]
X_train = X.ix[train_indices]# Select the rows
Y_train = Y.ix[train_indices]# Select the rows
X_test = X.ix[test_indices]  # Select the rows
Y_test = Y.ix[test_indices]  # Select the rows
print X_test.head()
print Y_test.head()
```

## Training and predictions

```python
# Initialize random forest

clf = ensemble.RandomForestClassifier(n_estimators = 900)
clf.fit(X_train, Y_train)
predictions = clf.predict(X_test)
sum(sum((Y_test - predictions)**2))
```

```python
n_range =  range(50, 1500, 50)
pred_error = []
MSE = 0
for n in n_range:
    clf = ensemble.RandomForestClassifier(n_estimators = n)
    clf.fit(X_train, Y_train)
    predictions = clf.predict(X_test)
    MSE = sum(sum((Y_test - predictions)**2))
    pred_error += [MSE]
plt.plot(n_range, pred_error)

    
```
