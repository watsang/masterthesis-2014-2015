```python
def create_sparse_matrices(stoichiometric_reactions, folder_links):    
    directory = "data_organism_reactions/" + "Sparse_Stoichiometric_matrices/"
    if not os.path.exists(directory):
        os.makedirs(directory)
    reactions = []
    compounds = []
    
    for file_adress, file_name in zip(folder_links, stoichiometric_reactions):
        S = pd.read_csv(file_adress, index_col=0)
        reactions += [S.columns.values]
        compounds += [S.index.values]
        S = np.int8(S.values)
        save_name = file_name.split(".txt")[0] + ".npy"
        S = np.save(directory +  save_name, S)

    dict_comp_react = pd.DataFrame([stoichiometric_reactions, reactions, compounds], index=["stoichiometric_matrix", "reactions", "compounds"])
    dict_comp_react = dict_comp_react.T
    dict_comp_react.to_csv(directory + "dict_comp_react.csv", index_col=0)
    return dict_comp_react
        
        
```

```python
from numba import autojit

def _calc_missing_compounds(compounds1, compounds2):
    """
    @param compounds1: compounds of organism1
    @param compounds2: compounds of organism2
    @return: returns missing compounds of organism1 and organism2
    """
    # cp_add1 = compounds2[~compounds2.isin(compounds1)]
    # cp_add2 = compounds1[~compounds1.isin(compounds2)]
    return compounds2[~compounds2.isin(compounds1)], compounds1[~compounds1.isin(compounds2)]

def create_frames(S1, S2, compounds1, compounds2):
    """
    @param S1: accepts sparse stoichiometric matrix 1
    @param S2: accepts sparse stoichiometric matrix 2
    @param compounds1: compounds of organism1
    @param compounds2: compounds of organism2
    @return: returns two stoichiometric matrices where the rows:
            * consist of the same compounds
            * are sorted according to the index
    """   
    # Decide which compounds to add
    cp_add1, cp_add2 = _calc_missing_compounds(compounds1, compounds2)
    
    n_add1 = len(cp_add1)
    n_add2 = len(cp_add2)
    
    n_compounds1 = len(compounds1)
    n_compounds2 = len(compounds2)
    
    # Initialize size of finale matrices
    frame1 = np.zeros([n_add1 + n_compounds1, S1.shape[1]]) # Create matrix to append matrix1
    frame2 = np.zeros([n_add2 + n_compounds2, S2.shape[1]]) # Create matrix to append matrix2
    
    # Assign values of stoichiometric matrices
    frame1[0:n_compounds1,:] = S1
    frame2[0:n_compounds2,:] = S2
    
    # Create frames to return    
    frame1 = pd.DataFrame(frame1, index = list(compounds1) + list(cp_add1))
    frame2 = pd.DataFrame(frame2, index = list(compounds2) + list(cp_add2))

    # Sort rowwise
    frame1 = frame1.sort_index()
    frame2 = frame2.sort_index()
    
    # return frame1.sort_index().values, frame2.sort_index().values
    return frame1.values, frame2.values

create_frames_numba = autojit()(create_frames)
```

```python
#import multiprocessing as mp
#from functools import partial
#from copy import deepcopy
# from joblib import Parallel, delayed
    
def CCA_matrix(list_of_S, list_of_compounds, list_of_names):
    """
    @param list_of_S: accepts a list of locations where the stoichiometric matrices can be loaded
    @return: returns a pd.DataFrame with all lambda values corresponding lambda values
    """
    # cpus = mp.cpu_count()
    # Start multiprocessing
    n = len(list_of_S)
    df_CF = np.zeros([n, n])
    for i in xrange(n):

        S1 = list_of_S[i]
        S1_compounds = list_of_compounds[i]

        row_list_of_S = [list_of_S[j + i] for j in xrange(n-i)]
        row_list_of_compounds = [list_of_compounds[j + i] for j in xrange(n-i)]

        df_CF[i, i:n] = [total_CCA(S1, x, S1_compounds, y) for x, y in zip(row_list_of_S, row_list_of_compounds)]
    for i in xrange(n):
        for j in xrange(i):
            df_CF[i, j] = df_CF[j, i]  
       
    return pd.DataFrame(df_CF, columns= list_of_names, index = list_of_names)

# CCA_matrix = autojit()(CCA_matrix)
```
