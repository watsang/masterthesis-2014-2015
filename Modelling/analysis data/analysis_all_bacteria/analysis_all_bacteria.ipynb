## 1) Calculate Jaccard based on compounds

```python
def load_all_compounds(list_of_S, data_folder):
    """
    @param: list of all stoichiometric matrices to be loaded
    @return: dictionary with all reactions for each bacteria
    """
    dictionary_reactions = {}
    for bacteria in list_of_S:
        location_file = "".join([data_folder, bacteria])
        dictionary_reactions[bacteria] = pd.read_csv(location_file, index_col=0).index
    return dictionary_reactions
```

```python
def jacard_index_compounds(v1, v2):
    """
    @param S1: vector 1 with all the compounds
    @param S2: vector 2 with all the compounds
    @return: Jacard index

    MATH description..? => only counts the number of shared locations
             |v1 ? v2|
    Jaccard = ---------
             |v1 U v2|
    """
    compounds_intersect = v1.isin(v2).sum()
    compounds_different1 = len(v1) - compounds_intersect
    compounds_different2 = len(v2) - compounds_intersect

    jacard_index = compounds_intersect
    union = (compounds_intersect + compounds_different1 + compounds_different2 + 0.)
    jacard_index /= union
    return jacard_index

def jacard_matrix_compounds(list_of_S, data_folder):
    """
    @param list_of_S: accepts a list from where the stoichiometric matrices can be loaded
    @param data_folder: string containing adress where all the stoichiometric matrices are stored
    @return: returns a pd.DataFrame with all lambda values corresponding jaccard values
    """
    n = len(list_of_S)
    df_jm = np.zeros([n, n])
    column = list_of_S

    dictionary_compounds = load_all_compounds(list_of_S, data_folder)
    for i in xrange(n):
        bacteria1 = list_of_S[i]
        vector1 = dictionary_compounds[bacteria1]
        for j in xrange(i + 1):
            bacteria2 = list_of_S[j]
            vector2 = dictionary_compounds[bacteria2]
            fill_value = jacard_index_compounds(vector1, vector2)
            df_jm[i,j] = fill_value
    for i in xrange(n):
        for j in xrange(n - 1 - i):
            df_jm[i, j + 1 + i] = df_jm[j + 1 + i, i]
    return pd.DataFrame(df_jm, columns= list_of_S, index =  list_of_S)
```

## 2) similarity Jaccard based on location

```python
def jacard_index_location(v1, v2):
    """
    @param S1: vector 1 with all the locations
    @param S2: vector 2 with all the locations
    @return: Jacard index
    
    MATH description..? => only counts the number of shared locations
             |v1 ∩ v2|
    Jacard = ---------
             |v1 U v2|
    """
    location_intersect = v1.isin(v2).sum()
    location_different1 = len(v1) - location_intersect
    location_different2 = len(v2) - location_intersect
    
    jacard_index = location_intersect
    union = (location_intersect + location_different1 + location_different2 + 0.)
    jacard_index /= union
    return jacard_index

def jacard_matrix_location(location_matrix):
    n = location_matrix.shape[0]
    df_jm = np.zeros([n, n])
    column = location_matrix.columns
    for i in xrange(n):
        for j in xrange(i + 1):
            vector1 = location_matrix.iloc[i,:]
            vector2 = location_matrix.iloc[j,:]
            vector1 = column[map(bool, vector1)]
            vector2 = column[map(bool, vector2)]
            df_jm[i,j] = jacard_index_location(vector1, vector2)
    for i in xrange(n):
        for j in xrange(n - 1 - i):
            df_jm[i, j + 1 + i] = df_jm[j + 1 + i, i]             
    return pd.DataFrame(df_jm, columns= location_matrix.index, index =  location_matrix.index)
    
    
```
