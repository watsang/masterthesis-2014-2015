## 1) Canonical Correlation Analysis

```python
def create_frames(S1, S2):
    """
    @param S1: accepts stoichiometric matrix 1
    @param S2: accepts stoichiometric matrix 2
    @return: returns two stoichiometric matrices where the rows:
            * consist of the same compounds
            * are sorted according to the index
    """
    # Determine which compounds to add
    cp_add_1 = S2.index[~S2.index.isin(S1.index)] # Find compounds not present in S1
    cp_add_2 = S1.index[~S1.index.isin(S2.index)] # Find compounds not present in S2
    
    # Create matrix you want to append
    append1 = np.zeros([len(cp_add_1), S1.shape[1]]) # Create matrix to append matrix1
    append2 = np.zeros([len(cp_add_2), S2.shape[1]]) # Create matrix to append matrix2
    
    # Create frames to return
    frame1 = S1.append(pd.DataFrame(append1, columns = S1.columns, index = cp_add_1))
    frame2 = S2.append(pd.DataFrame(append2, columns = S2.columns, index = cp_add_2))

    # Sort rowwise
    frame1 = frame1.sort_index()
    frame2 = frame2.sort_index()
    
    return frame1, frame2

def CCA(S1, S2):
    """
    @param S1: accepts stoichiometric matrix 1
    @param S2: accepts stoichiometric matrix 2
    @return: returns the lambda value of the canonical correlation analysis (CCA)
    
    CCA
    ---
    formula:
    [ (S1^T * S1 + I * lambda)^-1 * (S1^T * S2) * (S2^T * S2 + I * lambda)^-1 ] * (S2^T * S1) a = lambda^2 * a 
    
    implementation:
    term1 = (S1^T * S1 + I * lambda)^-1 
    term2 = (S1^T * S2)
    term3 = (S2^T * S2 + I * lambda)^-1 
    term4 = (S2^T * S1)
    lambda is a measure of correlation between two stoichiometric matrices
    
    """
    # Create matrices with the same compounds
    S1, S2 = create_frames(S1, S2)
      
    # Calculate CCA-measure
    term1 = np.dot(S1.transpose(), S1)
    term1 = np.linalg.inv(term1 + np.eye(term1.shape[0]) * .001)
    
    term2 = np.dot(S1.transpose(), S2)
    
    term3 = np.dot(S2.transpose(), S2)
    term3 = np.linalg.inv(term3 + np.eye(term3.shape[0]) * .001)
    
    term4 = np.dot(S2.transpose(), S1)
    
    lambda_square = np.dot(term1, term2)
    lambda_square = np.dot(lambda_square, term3)
    lambda_square = np.dot(lambda_square, term4)
    #lambda_square = np.around(lambda_square, decimals = 3)
    # return np.sqrt(max(real(np.linalg.eig(lambda_square)[0])))
    return np.trace(lambda_square)


def CCA_matrix(list_of_S):
    """
    @param list_of_S: accepts a list of Stoichiometric Matrices
    @return: returns a pd.DataFrame with all lambda values corresponding lambda values
    """
    n = len(list_of_S)
    df_CF = np.zeros([n, n])
    for i in xrange(n):
        for j in xrange(i + 1):
            S1 = eval(list_of_S[i])
            S2 = eval(list_of_S[j])
            df_CF[i,j] = CCA(S1, S2)
    for i in xrange(n):
        for j in xrange(n - 1 - i):
            df_CF[i, j + 1 + i] = df_CF[j + 1 + i, i]  
       
    return pd.DataFrame(df_CF, columns= list_of_S, index = list_of_S)
```

## 2) Jacard-index (a): based on reaction

```python
def jacard_index(S1, S2):
    """
    @param S1: stoichiometric matrix not "equalized" with S2
    @param S2: stoichiometric matrix not "equalized" with S1
    @return: Jacard index
    
    MATH description..? => only counts the number of shared reactions
             |S1 ∩ S2|
    Jacard = ---------
             |S1 U S2|
    """
    reaction_intersect = S2.columns.isin(S1.columns).sum()
    reaction_different1 = S1.shape[1] - reaction_intersect
    reaction_different2 = S2.shape[1] - reaction_intersect
    
    jacard_index = reaction_intersect 
    union = (reaction_intersect + reaction_different1 + reaction_different2 + 0.)
    jacard_index /= union
    return jacard_index

def jacard_matrix(list_of_S):
    n = len(list_of_S)
    df_jm = np.zeros([n, n])
    for i in xrange(n):
        for j in xrange(n):
            S1 = eval(list_of_S[i])
            S2 = eval(list_of_S[j])
            df_jm[i,j] = jacard_index(S1, S2)
    return pd.DataFrame(df_jm, columns= list_of_S, index = list_of_S)
```

## 2) Jacard-index (b): based on compounds

```python
def jacard_index_compounds(S1, S2):
    """
    @param S1: stoichiometric matrix not "equalized" with S2
    @param S2: stoichiometric matrix not "equalized" with S1
    @return: Jacard index
    
    MATH description..? => only counts the number of shared reactions
             |S1 ∩ S2|
    Jacard = ---------
             |S1 U S2|
    """
    compounds_intersect = S2.index.isin(S1.index).sum()
    compounds_different1 = S1.shape[0] - compounds_intersect
    compounds_different2 = S2.shape[0] - compounds_intersect
    
    jacard_index = compounds_intersect 
    union = (compounds_intersect + compounds_different1 + compounds_different2 + 0.)
    jacard_index /= union
    return jacard_index

def jacard_matrix_compounds(list_of_S):
    n = len(list_of_S)
    df_jm = np.zeros([n, n])
    for i in xrange(n):
        for j in xrange(n):
            S1 = eval(list_of_S[i])
            S2 = eval(list_of_S[j])
            df_jm[i,j] = jacard_index_compounds(S1, S2)
    return pd.DataFrame(df_jm, columns= list_of_S, index = list_of_S)
```

```python
def find_name(text):
    """
    Find full name in database and return that value
    """
    selection = names_bacteria['file_name'].isin([text])
    return names_bacteria[selection].full_name.values[0]

full_name = map(lambda x: find_name(x), names)
full_name
```

```python
loc_matrix_narrow = location_matrix[location_matrix.index.isin(full_name)]
loc_matrix_narrow = loc_matrix_narrow.sort()
loc_matrix_narrow = loc_matrix_narrow.drop_duplicates(cols='index', take_last=True)
```

```python
def jacard_index_location(v1, v2):
    """
    @param S1: vector 1 with all the locations
    @param S2: vector 2 with all the locations
    @return: Jacard index
    
    MATH description..? => only counts the number of shared locations
             |v1 ∩ v2|
    Jacard = ---------
             |v1 U v2|
    """
    location_intersect = v1.isin(v2).sum()
    location_different1 = len(v1) - location_intersect
    location_different2 = len(v2) - location_intersect
    
    jacard_index = location_intersect
    union = (location_intersect + location_different1 + location_different2 + 0.)
    jacard_index /= union
    return jacard_index

def jacard_matrix_location(location_matrix):
    n = location_matrix.shape[0]
    df_jm = np.zeros([n, n])
    column = location_matrix.columns
    for i in xrange(n):
        for j in xrange(i + 1):
            vector1 = location_matrix.iloc[i,:]
            vector2 = location_matrix.iloc[j,:]
            vector1 = column[map(bool, vector1)]
            vector2 = column[map(bool, vector2)]
            df_jm[i,j] = jacard_index_location(vector1, vector2)
    for i in xrange(n):
        for j in xrange(n - 1 - i):
            df_jm[i, j + 1 + i] = df_jm[j + 1 + i, i]             
    return pd.DataFrame(df_jm, columns= location_matrix.index, index =  location_matrix.index)
    
    
```
