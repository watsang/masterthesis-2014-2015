```python
organism_names['first_name'] = map(lambda x: x.split(" ")[0], organism_names['full_name'])
organism_names['second_name'] = map(lambda x: x.split(" ")[1], organism_names['full_name'])
organism_names['abbreviated_name'] = map(lambda x: " ".join([organism_names['first_name'][x], organism_names['second_name'][x]]), xrange(organism_names.shape[0]))
```

```python
# Filter for all the organisms that have labels for the locations
def check_name(search_key):
    """
    Checks if name is substring in the list of organism_name of processed_names
    """
    return any([search_key in name for name in processed_names['organism_name']])

selection = map(lambda x: check_name(x), organism_names['first_name'])
organism_names = organism_names[selection]
organism_names.index = range(organism_names.shape[0])
```

```python
# Add locations for each organism
def find_locations(search_key):
    """
    @param serach_key: the first name of the bacteria
    @return: list with True or False on the corresponding row where
        the search_key is a substring of a string in processed_names
    """
    selection = [search_key in name for name in processed_names['organism_name']]
    return ";".join(processed_names[selection].extracted_keyword.unique())

organism_names['locations'] = map(lambda x: find_locations(x), organism_names['first_name'])
```
