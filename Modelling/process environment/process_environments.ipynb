```python
folder_data = "gg_otus_allseqs"
```

```python
files  = os.listdir(folder_data)
files
```

```python
text_files = filter(lambda x: "txt" in x, files)
text_files = map(lambda x: "/".join([folder_data, x]), text_files)
text_files
```

The fasta description of each sequence is as follow: GI | Greengenes_ID | sequence_type | organism_name | NCBI_taxonomy | RDP_taxonomy | Hugenholtz_taxonomy

```python
import fileinput

environment_file = '/'.join([folder_data, 'bacteria_environment.txt'])
f = open(environment_file,'w')

for file_name in text_files:   
    for line in fileinput.input([file_name]):
        if '>' in line:
            elements = line.split(' | ')
            name = elements[3]
            # if the organism is uncultured don't write it in the file
            #if 'uncultured' not in name:
            # NCBI_taxonomy | RDP_taxonomy | Hugenholtz_taxonomy
            NCBI_taxonomy = elements[4]
            RDP_taxonomy = elements[5]
            Hugenholtz_taxonomy = elements[6]
            write = '\t'.join([elements[1], name, NCBI_taxonomy, RDP_taxonomy, Hugenholtz_taxonomy]) + '\n'
            f.write(write) # python will convert \n to os.linesep
    
f.close() # you can omit in most cases as the destructor will call if
```

```python
# Column heads are Greengenes_ID, ENVO_ID, extracted_keyword, sequence_isolation_source
otu_columns = ['Greengenes_ID', 'ENVO_ID', 'extracted_keyword', 'sequence_isolation_source']
otu_environment = pd.read_csv('/'.join([folder_data, 'gg_allseqs2envo.tsv']), sep= '\t', header=0, names = otu_columns)
otu_environment
```

```python
# Filter the processed names for only the green_genes_id that are present in the database
processed_names = folder_data + '/' + 'bacteria_environment.txt'
# NCBI_taxonomy | RDP_taxonomy | Hugenholtz_taxonomy
processed_names = pd.read_csv(processed_names, header = 0, names = ['Greengenes_ID', 'organism_name', 'NCBI_taxonomy', 'RDP_taxonomy', 'Hugenholtz_taxonomy'], sep = '\t')
selection = processed_names['Greengenes_ID'].isin(otu_environment['Greengenes_ID'])
processed_names = processed_names.ix[selection,:]
processed_names.index = range(processed_names.shape[0])
```

```python
processed_names
```

```python
# find all the associated keywords with the Greengenes_ID and sequence_isolation_source and add them to each row
processed_names['extracted_keyword'] = [otu_environment[processed_names['Greengenes_ID'][i] == otu_environment['Greengenes_ID']]['extracted_keyword'].values[0] for i in xrange(processed_names.shape[0])]
processed_names['sequence_isolation_source'] = [otu_environment[processed_names['Greengenes_ID'][i] == otu_environment['Greengenes_ID']]['sequence_isolation_source'].values[0] for i in xrange(processed_names.shape[0])]

```

```python
processed_names['extracted_keyword'] = [word.lower() for word in processed_names['extracted_keyword']]
```
